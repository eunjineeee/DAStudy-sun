## 구매 비중 확인하기

* USA, Ireland, UK 순으로 구매비중이 형성
* 신용카드 사용 비중이 반반인 것으로 드러남
    * 한국에서는 현금 대신 카드를 많이 사용하는 것에 비해 달러, 유로 등 카드 대신 화폐를 많이 쓰는 것으로 추정

```python
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
%matplotlib inline

colors = ['#EB3B61', '#893BEB', '#87EBE2']

plt.figure(figsize = (14, 8))
plt.pie(x=customer['Country'].value_counts(normalize=True), labels=customer['Country'].unique(), autopct='%1.1f%%', colors=colors)
plt.show()
```

## 손님들의 구매 수량 확인하기

* 손님이 한 번에 커피를 사갈 때 몇 잔을 사갈까 ?
* 1잔 ~ 6잔 까지 비등비등한 배분을 보인다.
    * 한국은 1잔 or 2잔 비중이 매우 높고, 나머지는 낮을 것 같은데.. 해외는 다른가보다..

```python
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
%matplotlib inline

sns.countplot(data = df_total, x=df_total.Quantity)
plt.xlabel('Quantity')
plt.ylabel('Count of Customers')
plt.title('Quantity of the coffee product ordered')
plt.show()
```

* Robusta, Exc, Arabica, Lib Coffee
* 커피 종류에 따른 구매 수량 차이 확인

```python
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
%matplotlib inline

plt.figure(figsize=(15, 6))
sns.countplot(data = df_total, x=df_total.Quantity, hue ='Coffee Type')
plt.xlabel('Quantity')
plt.ylabel('Count of Customers')
plt.title('Quantity of the coffee product ordered')
plt.legend(bbox_to_anchor = (1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
